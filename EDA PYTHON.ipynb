```python
# Total Accidents
total_accidents = accidents['Accident_ID'].nunique()

# Total Injuries, Fatalities, Casualties
total_injuries = accidents['Number of Injuries'].sum()
total_fatalities = accidents['Number of Fatalities'].sum()
total_casualties = total_injuries + total_fatalities

# Average Rates
avg_fatality_rate = total_fatalities / total_casualties
avg_injury_rate = total_injuries / total_casualties

# Top 5 Accident Causes
top_causes = accidents['Accident Cause'].value_counts().head(5)

# Accidents per Country
accidents_per_country = accidents['Country'].value_counts()

# Accidents per Month
accidents_per_month = accidents['Month'].value_counts()

# Accidents per Day of Week
accidents_per_day = accidents['Day of Week'].value_counts()

# Accidents per Time of Day
accidents_per_time = accidents['time of day'].value_counts()

# Weather Conditions
accidents_per_weather = weather['Weather Conditions'].value_counts()

# Road Conditions
accidents_per_road = weather['Road Condition'].value_counts()

# Display results
print(f"Total Accidents: {total_accidents}\n")
print(f"Total Casualties: {total_casualties}")
print(f"Total Injuries: {total_injuries}")
print(f"Total Fatalities: {total_fatalities}")
print(f"Average Fatality Rate: {avg_fatality_rate:.6f}")
print(f"Average Injury Rate: {avg_injury_rate:.6f}\n")

print("Top 5 Accident Causes:")
print(top_causes, "\n")

print("Accidents per Country:")
print(accidents_per_country, "\n")

print("Accidents per Month:")
print(accidents_per_month, "\n")

print("Accidents per Day of Week:")
print(accidents_per_day, "\n")

print("Accidents per Time of Day:")
print(accidents_per_time, "\n")

print("Accidents per Weather Condition:")
print(accidents_per_weather, "\n")

print("Accidents per Road Condition:")
print(accidents_per_road)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted")

# ------------------------
# Dashboard
# ------------------------
fig, axes = plt.subplots(4, 2, figsize=(18, 20))
fig.suptitle("Traffic Accident Dashboard", fontsize=22)

# 1   
sns.countplot(y='Country', data=accidents,
              order=accidents['Country'].value_counts().index, ax=axes[0,0])
axes[0,0].set_title("Accidents per Country")

# 2   
sns.countplot(y='Month', data=accidents,
              order=accidents['Month'].value_counts().index, ax=axes[0,1])
axes[0,1].set_title("Accidents per Month")

# 3   
sns.countplot(y='Day of Week', data=accidents,
              order=accidents['Day of Week'].value_counts().index, ax=axes[1,0])
axes[1,0].set_title("Accidents per Day of Week")

# 4    
sns.countplot(y='time of day', data=accidents,
              order=accidents['time of day'].value_counts().index, ax=axes[1,1])
axes[1,1].set_title("Accidents per Time of Day")

# 5    
sns.countplot(y='Weather Conditions', data=weather,
              order=weather['Weather Conditions'].value_counts().index, ax=axes[2,0])
axes[2,0].set_title("Accidents per Weather Condition")

# 6    
sns.countplot(y='Road Condition', data=weather,
              order=weather['Road Condition'].value_counts().index, ax=axes[2,1])
axes[2,1].set_title("Accidents per Road Condition")

# 7  
sns.countplot(y='Vehicle Condition', data=vehicles,
              order=vehicles['Vehicle Condition'].value_counts().index, ax=axes[3,0])
axes[3,0].set_title("Vehicle Condition Distribution")

# 8    
sns.histplot(locations['Population Density'], bins=20, kde=True, ax=axes[3,1])
axes[3,1].set_title("Population Density Distribution")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
